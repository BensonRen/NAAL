```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bs100_folder = '/home/sr365/NAAL/results/ADM_num_layer_8_nuron_1500_nmod_10_toMSE_0.002_dx_100_pool_mul_5_naal_True'
bs100_subfolder = 'ADM_Random_retrain_False_complexity_8_bs_500_pool_500_dx_100_step_1_x0_500_nmod_10_trail_0'
mse_file = 'test_mse.npy'
```

```python
f = plt.figure(figsize=[8,4])
# Plot the different batch size results
for bs in [100, 200, 500, 1000, 2000]:
    # Get the fulll path of the file
    mse_full_path = os.path.join(bs100_folder, bs100_subfolder.replace('bs_500', 'bs_{}'.format(bs)), mse_file)
    # Read in MSE
    mse = np.load(mse_full_path)
    plt.plot(mse, label='bs_{}'.format(bs))
plt.legend()
plt.xlabel('AL step')
plt.ylabel('mse')
plt.yscale('log')
plt.title('Batch_size effect for ')
```

```python
bs100_xlabel = np.arange(100) * 100 + 500
```

```python
f = plt.figure(figsize=[8,4])
# Plot the different batch size results
for bs in [100, 200, 500, 1000, 2000]:
    # Get the fulll path of the file
    mse_full_path = os.path.join(bs100_folder, bs100_subfolder.replace('bs_500', 'bs_{}'.format(bs)), mse_file)
    # Read in MSE
    mse = np.load(mse_full_path)
    plt.plot(bs100_xlabel[:len(mse)], mse, label='bs_{}'.format(bs))
plt.legend()
plt.xlabel('AL step')
plt.ylabel('mse')
plt.yscale('log')
plt.title('Batch_size effect for AL step of 100 points')
```

```python
bs50_folder = '/home/sr365/NAAL/results/ADM_num_layer_8_nuron_1500_nmod_10_toMSE_0.002_dx_50_pool_mul_5_naal_True'
bs50_subfolder = 'ADM_Random_retrain_False_complexity_8_bs_500_pool_250_dx_50_step_1_x0_500_nmod_10_trail_0'
mse_file = 'test_mse.npy'
bs50_xlabel = np.arange(100) * 50 + 500
```

## Comparing across BS

```python
f = plt.figure(figsize=[8,4])
# Plot the different batch size results
for bs in [200, 500, 1000]:
    # Get the fulll path of the file
    bs50_mse_full_path = os.path.join(bs50_folder, bs50_subfolder.replace('bs_500', 'bs_{}'.format(bs)), mse_file)
    bs100_mse_full_path = os.path.join(bs100_folder, bs100_subfolder.replace('bs_500', 'bs_{}'.format(bs)), mse_file)
    # Read in MSE
    bs50_mse = np.load(bs50_mse_full_path)
    bs100_mse = np.load(bs100_mse_full_path)
    plt.plot(bs50_xlabel[:len(bs50_mse)], bs50_mse, '--', label='dx50_bs_{}'.format(bs))
    plt.plot(bs100_xlabel[:len(bs100_mse)], bs100_mse, label='dx_100_bs_{}'.format(bs))
plt.legend()
plt.xlabel('AL step')
plt.ylabel('mse')
plt.yscale('log')
plt.title('Batch_size effect for AL step of 100 points')
```
